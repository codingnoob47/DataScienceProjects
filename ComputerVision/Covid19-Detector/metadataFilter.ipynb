```python
import pandas as pd
```

```python
def member_identification():
  path = '/content/gdrive/MyDrive/ComputerVision/Covid19-Detector/'
  metadata = pd.read_csv(path + 'metadata.csv')
  metadata = metadata[(metadata['finding'] == 'Pneumonia/Viral/COVID-19') & (metadata['view'] == 'AP') & (metadata['RT_PCR_positive'] == 'Y')]
  print("Total Unique Members :  ", metadata['patientid'].nunique())  

  return metadata
```

```python
metadata = member_identification()
```

```python
def get_patient_information(metadata):
  patient_id_filename_information = metadata[['patientid','filename']].drop_duplicates()
  patient_id_filename_information.to_csv(path + 'patient_id_filename_information.csv')

  return patient_id_filename_information
  
```

```python
patient_id_filename_information = get_patient_information(metadata)
```

```python
import cv2
```

```python
class SimplePreprocessor:

  def __init__(self, width, height, inter = cv2.INTER_AREA):

    self.width = width
    self.height =  height
    self.inter = inter

  def preprocess(self, image):

    return cv.resize(image, (self.width, self.height), interpolation = self.inter)
```
